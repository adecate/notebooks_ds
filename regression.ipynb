```
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(train.shape, test.shape)
```

```
from sklearn.ensemble import ExtraTreesRegressor
```

```
X_train = train.drop(columns = ['critical_temperature'])
y_train = train['critical_temperature']

X_test = test.copy()

print(X_train.shape)
print(y_train.shape)
```

```
X_tt, X_tv, y_tt, y_tv = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```
%time
reg = ExtraTreesRegressor(n_estimators=100, random_state=0)
reg.fit(X_train, y_train)
print(reg.score(X_tt, y_tt))
```

```
et_fin = reg.predict(X_test)
```

```
et_fin = pd.Series(et_fin).round(2)
et_fin
```

```
et_fin.to_csv('answer.csv',index=False, header = None)
```
